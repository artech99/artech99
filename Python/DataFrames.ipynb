```python
import pandas as pd
import numpy as np
import random
```

```python
num_rows = 100
colors = ['Red', 'Blue', 'Green']

df = pd.DataFrame( {'color': [colors[random.randint(0,2)] for _ in range(num_rows)],
                    'integers': [random.randint(0,15) for _ in range(num_rows)],
                    'floats': [random.random() for _ in range(num_rows)]})
```

```python
# return data for rows 10 thru 13 for the 'color' and 'integer' columns

df.loc[10:13, ['color','integers']]
```

```python
# return data for rows 10 thru 13 and the first 3 columns

df.iloc[10:13, 0:2]
```

```python
# return 'True' for all rows where the color is equal to 'Red' and add it to a data frame

red_mask = df.loc[:,'color'] == 'Red'

red_df = df[red_mask]
red_df
```

```python
data = { 'even': range(20,0,-2),
         'odd':  range(1,21,2)
       }
df = pd.DataFrame(data)
df

# return the maximun valuemaximum value of each column

df.apply(max)
```

```python
data = { 'even': range(20,0,-2),
         'odd':  range(1,21,2)
       }
df = pd.DataFrame(data)
df

# return the maximun valuemaximum value of each row

df.apply(max, axis=1)
```

```python
# create a function to compare whether value of odd or even number is larger

def odd_bigger(row):
    if row['odd'] > row['even']:
        return True
    return False


# apply the function to each row in the DataFrame

df.apply(odd_bigger, axis=1)


# add the results to a new column in the DataFrame

df['Odd Bigger'] = df.apply(odd_bigger, axis=1)
df
```
